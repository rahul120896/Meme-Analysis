```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url="https://codecodehub.blogspot.com/2019/04/top-programming-memes.html"
```

```python
page=req.get(url)
```

```python
print(page.text)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
link=soup.findAll('img')
```

```python
for tag in link:
    print(tag['src'])
```
